Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and preProcessing

```python
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

Selecting the relevant features for recommendation

```python
selected_features = ['genres','keywords','tagline','cast','director']
```

```python
print(selected_features)
```

```python
#replacing the null values with NULL string
#it takes each coloumn in the selected features and replaces all the missing values with null string

for features in selected_features:
  movies_data[features] = movies_data[features].fillna('')
```

```python
#combining all the 5 selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
#coverting text data into numerical values which is feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Finding similarity score using cosine similarity algorithm


```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#the above output means that each movie is compared with all the 4803 other movies
```

```python
#taking user input and then find if its in the dataset and then find movies similar to it using the similarity score
movie_name = input("Enter the name of the your favorite movie!")
```

```python
#creating a list with all the movies and names given in the dataset
movie_list = movies_data['title'].tolist()
print(movie_list)
```

```python
#finding best match of the value given by the user
find_close_match = difflib.get_close_matches(movie_name, movie_list)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#find the index of the movie with title
index_of_movie = movies_data[movies_data.title == close_match]['index'].iloc[0] # Use .iloc[0] to access the first element of the Series
print(index_of_movie)
```

```python
#getting list of similar movies

similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sort based on similarity score values
#high to low similarity value is given by reverse(descending order)
#lambda x:x[1] means we are taking the first value in the list and not the 0th element
sorted_list = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_list)
```

```python
#print the name of similar movies based on index
print('Movies Suggested for you: \n')
i=1
for movie in sorted_list:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if i<11:
    print(i,'.',title_from_index)
    i=i+1

```

Popularity Based Movie Recommendation

```python
#popularity alreayd given in numerical values

popularity_with_index = list(enumerate(movies_data['popularity']))
sorted_popularity = sorted(popularity_with_index, key=lambda item: item[1], reverse=True)
print(sorted_popularity)
```

```python
print("Most popular movies are: \n")
i = 1
for movie in sorted_popularity:
  ind = movie[0]
  title_from_index = movies_data[movies_data.index == ind]['title'].values[0]
  if i<10:
    print(i,'.',title_from_index)
    i=i+1
```
